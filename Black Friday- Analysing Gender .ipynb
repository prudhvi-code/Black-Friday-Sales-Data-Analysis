```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("BlackFriday.csv")
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
for i in df.columns:
    print(i,':',df[i].nunique())
```

```python
len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])
```

```python
data=pd.DataFrame({'Ratio':[len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])]},index=['Male','Female'])
```

```python
data
```

```python
data.plot.pie(y='Ratio',figsize=(6,6),autopct='%.2f')
```

```python
df.groupby('Gender').size().plot(kind='pie',autopct='%.2f',title='Gender Ratio',figsize=(6,6))
```

```python
df.groupby('Gender').size().plot(kind='bar',title='Gender Ratio',figsize=(6,6))
```

```python
df.groupby('Gender').agg({'Purchase': 'sum'})
```

```python
df.groupby('Gender').agg({'Purchase': 'sum'}).plot(
    kind='pie', y='Purchase', autopct='%1.1f%%', startangle=90, figsize=(6,6)
)
```

```python
df.groupby('Gender').agg({'Purchase': 'mean'})
```

```python
df.groupby('Gender').agg({'Purchase': 'mean'}).plot(
    kind='pie', y='Purchase', autopct='%1.1f%%', startangle=90, figsize=(6,6)
)
```
